```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load the model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3.5-mini-instruct",
    device_map="cuda",  # Ensure the model is loaded onto the GPU
    torch_dtype="auto",  # Automatically select the appropriate torch dtype
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3.5-mini-instruct")

# Initialize the text generation pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)
```

```python
from google.colab import drive
import json

# Mount Google Drive
drive.mount('/content/drive')

# Load the example JSON file
json_file_path = '/content/drive/MyDrive/My/FinalAirflowJSON.json'
with open(json_file_path, 'r') as file:
    example_json = json.load(file)

print("Example JSON loaded:")
print(json.dumps(example_json, indent=4))
```

```python
def collect_user_inputs():
    source_type = input("Enter source data source type: ")
    source_table_name = input("Enter source table name: ")
    source_table_join_key = input("Enter source table join key: ")
    destination_type = input("Enter destination data source type: ")
    destination_table_name = input("Enter destination table name: ")
    destination_join_key = input("Enter destination table join key: ")
    other_metadata = input("Enter any other metadata (leave blank if none): ")

    user_inputs = {
        "source_type": source_type,
        "source_table_name": source_table_name,
        "source_table_join_key": source_table_join_key,
        "destination_type": destination_type,
        "destination_table_name": destination_table_name,
        "destination_join_key": destination_join_key,
        "other_metadata": other_metadata
    }

    return user_inputs

user_inputs = collect_user_inputs()

# Construct the prompt based on user inputs and example JSON
prompt = (
    "<|system|>\nYou are a helpful assistant.<|end|>\n"
    "<|user|>\nPlease generate a JSON configuration for an ETL pipeline.\n\n"
    f"Source Data Source Type: {user_inputs['source_type']}\n"
    f"Source Table Name: {user_inputs['source_table_name']}\n"
    f"Source Table Join Key: {user_inputs['source_table_join_key']}\n"
    f"Destination Data Source Type: {user_inputs['destination_type']}\n"
    f"Destination Table Name: {user_inputs['destination_table_name']}\n"
    f"Destination Join Key: {user_inputs['destination_join_key']}\n"
    f"Other Metadata: {user_inputs['other_metadata']}\n"
    "\nUse the following JSON format as a reference:\n\n"
    f"{json.dumps(example_json, indent=4)}\n<|end|>\n"
    "<|assistant|>"
)

# Generate the JSON configuration
generation_args = {
    "max_new_tokens": 1000,
    "return_full_text": False,
    "temperature": 0.7,
    "do_sample": True,
    "top_k": 50,
    "top_p": 0.95,
}

output = pipe(prompt, **generation_args)
generated_json = output[0]['generated_text']

print("Generated JSON Configuration:\n")
print(generated_json)
```
